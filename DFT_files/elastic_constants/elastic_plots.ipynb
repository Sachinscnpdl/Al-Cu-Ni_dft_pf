```python
def plot_volume(df, name='plt_name'):
    
    import pandas as pd
    import matplotlib.pyplot as plt
    from matplotlib.font_manager import FontProperties
    from matplotlib.ticker import MultipleLocator
    from matplotlib.ticker import AutoMinorLocator

    
    
    # Select specific columns T, B, C11, C12, C44
    selected_columns = ['T', 'vol']
    df_selected = df[selected_columns]

    # Convert Kbar to MPa by dividing by 10. 
    # Ther results from QE thermo_pw DFT is obtained in Kbar unit
#     df_selected.loc[:, ['T', 'vol']] = df_selected.loc[:, ['B', 'C11', 'C12', 'C44']] / 10

    # Plotting
    plt.figure(figsize=(5, 5))

    plt.plot(df_selected['T'], df_selected['vol'], label=' ', linestyle='-', marker=' ', color='k', linewidth=2.5)
#     plt.plot(df_selected['T'], df_selected['C12'], label='C12', linestyle='--', marker=' ', color='k', linewidth=2.5)
#     plt.plot(df_selected['T'], df_selected['C44'], label='C44', linestyle='-.', marker=' ', color='k', linewidth=2.5)

    plt.xlabel('T (K)', fontsize=18, fontweight='bold')
    plt.ylabel(r'$\mathbf{Volume \ (au)^3}$', fontsize=18, fontweight='bold')
    plt.title(name, fontsize=16)

#     # Create a FontProperties object for the legend font
#     legend_font = FontProperties()
#     legend_font.set_size(14)
#     legend_font.set_weight('bold')
#     # Apply the custom font style to the legend
#     plt.legend(fontsize=12, prop=legend_font)
    
    plt.grid(False)  # Turn off the grid lines
    plt.tick_params(labelsize=16, width=2, length=8, direction='out', which='both')  # Customize axis tick parameters

    fontprops = FontProperties(size=16, weight='bold')
    plt.xticks(fontproperties=fontprops, color='indigo')
    plt.yticks(fontproperties=fontprops, color='indigo')

    # Add minor ticks
    # Add minor ticks
    plt.gca().xaxis.set_minor_locator(AutoMinorLocator(2))
    plt.gca().yaxis.set_minor_locator(AutoMinorLocator(2))
    plt.tick_params(which='minor', width=1, length=5)  # Adjust the width as needed

    # Set x-axis limits to start from 200
    plt.xlim(-10, plt.xlim()[1]+10)

    plt.tight_layout()
    plt.savefig(name+'_plot.png')
    plt.show()
    
    return plt
```
