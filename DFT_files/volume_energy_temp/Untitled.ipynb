```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MaxNLocator

# Load data from the CSV file (al2cu_vt.csv)
file = 'al2cu_vt.csv'
df = pd.read_csv(file)

# Create a figure and three axes
fig, ax1 = plt.subplots(figsize=(4.5, 5))

# First plot (energy vs volume)
color = 'red'
ax1.set_xlabel(r'Volume (a.u.)$\mathbf{^3}$', fontsize=18, fontweight='bold')
ax1.set_ylabel(r'Energy (Ry)', color=color, fontsize=18, fontweight='bold')
ax1.plot(df['volume'], df['energy'], color=color, label='E vs V', linewidth=2.5,linestyle=' ', 
         marker='*', markersize=10, markevery=12,)
ax1.tick_params(axis='y', labelcolor=color, labelsize=18, width=2.5,length=6, direction='in')
ax1.tick_params(axis='x', labelcolor=color, labelsize=18, width=2.5,length=6, direction='in')
ax1.yaxis.label.set_fontweight('bold')  # Set font weight for the left y-axis label

# Set the number of ticks to 4 for both x and y-axes
ax1.xaxis.set_major_locator(MaxNLocator(nbins=4))
ax1.yaxis.set_major_locator(MaxNLocator(nbins=4))

# Set fonts for left y-axis and x-axis tick labels
fontprops = {'size': 18, 'weight': 'bold'}
for label in (ax1.get_xticklabels() + ax1.get_yticklabels()):
    label.set_fontproperties(fontprops)

# Make the left y-axis tick labels bold
for label in ax1.get_yticklabels():
    label.set_fontweight('bold')

# Create a second y-axis
ax2 = ax1.twinx()

# Second plot (T vs volume)
color = 'indigo'
ax2.set_ylabel('T (K)', color=color, fontsize=18, fontweight='bold')
ax2.plot(df['volume'], df['T'], color=color, linestyle='-', label='T vs V', linewidth=2.5)
ax2.tick_params(axis='y', labelcolor=color, labelsize=18, width=2.5,length=6, direction='in')
ax2.yaxis.label.set_fontweight('bold')  # Set font weight for the right y-axis label

# Set the number of ticks to 4 for the right y-axis
ax2.yaxis.set_major_locator(MaxNLocator(nbins=7))

# Make the right y-axis tick labels bold
for label in ax2.get_yticklabels():
    label.set_fontweight('bold')

# Adding legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

# Create a FontProperties object for the legend font
legend_font = FontProperties()
legend_font.set_size(16)
legend_font.set_weight('bold')
ax1.legend(lines1 + lines2, labels1 + labels2, prop=legend_font, loc='upper left', bbox_to_anchor=(0.275,0.3), fontsize=12)

# Title for the entire plot
plt.title('al2cu \n', fontsize=18, fontweight='bold')




# Save the plot as a PNG image
plt.savefig('vet_al2cu.png', bbox_inches='tight', dpi=1200)

# Display the plot
plt.show()
```
