```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MultipleLocator
from matplotlib.ticker import AutoMinorLocator

# Load the data from the CSV file
data = pd.read_csv('dos.csv')

# Extract the data for Al and Al2Cu
al_fr = data['al_fr']
al_dos = data['al_dos']
alg_fr = data['alg_fr']
alg_dos = data['alg_dos']
al2cu_fr = data['al2cu_fr']
al2cu_dos = data['al2cu_dos']

# Create a single plot with two lines for Al and Al2Cu
plt.figure(figsize=(7, 4))
plt.plot(al_fr, al_dos, label='Al DOS', color='blue', linestyle='-', linewidth=2)
#plt.plot(alg_fr, alg_dos, label='Alg DOS', color='blue', linestyle='--', linewidth=2)
plt.plot(al2cu_fr, al2cu_dos, label=r'Al$\mathbf{_2}$Cu DOS', color='red', linestyle='-', linewidth=2)

# Add labels and a legend
plt.xlabel(r'Frequency (cm$^\mathbf{-1}$)', fontsize=22, fontweight='bold')
plt.ylabel(r'DOS (States / cm$\mathbf{^{-1}}$ /cell)', fontsize=22, fontweight='bold')
plt.title(' ', fontsize=22, fontweight='bold')
plt.legend()

plt.grid(False)  # Turn off the grid lines
plt.tick_params(labelsize=18, width=2, length=8, direction='in', which='both')  # Customize axis tick parameters

fontprops = FontProperties(size=18, weight=550)
plt.xticks(fontproperties=fontprops, color='k')
plt.yticks(fontproperties=fontprops, color='k')

# Add minor ticks
# Add minor ticks
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(2))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(2))
plt.tick_params(which='minor', direction='in', width=1, length=5)  # Adjust the width as needed

# Create a FontProperties object for the legend font
legend_font = FontProperties()
legend_font.set_size(14)
legend_font.set_weight('bold')
# Apply the custom font style to the legend
plt.legend(fontsize=16, prop=legend_font)

# Show the plot
#plt.grid()
plt.show()
```
